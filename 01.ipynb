## Equação da reta
A equação da reta é aprendida na forma ax+by+c, mas para nos adequarmos às nomenclaturas de redes neurais, podemos reescrever essa equação como **w1x1 + w2x2 + b**. Ou seja **w1**, **w2** e **b** são os parametros que definem o nosso modelo linear, a nossa reta.

Vamos começar realizando o plot de uma reta especifica **-1x¹ + 4x² + 0.4**

OBS: o simbolo '1' e '2' quee se assemelha uma potencia é apenas o index da variavel.


````python
import numpy as np

a = -1
b = 4
c = 0.4

#ax + by + c = 0

x = np.linspace(-2, 4, 50)

# para descobrirmos o valor de y devemos islolar o mesmo

y = (-a*x - c)/b

print('x:')
print(x)
print()
print('y:')
print(y)
````

````python
#Plotando uma reta com os valores obtidos
import matplotlib.pyplot as plt

def plot_line(a,b,c,x):
  y = (-a*x -c)/b
  plt.axvline(0, -1,1,color='k', linewidth=1)
  plt.axhline(0, -2,4,color='k', linewidth=1)
  plt.plot(x,y)
  plt.grid(True)

plot_line(a,b,c,x)
````

## Solucionando a equação da reta para diferentes pontos
A seguir vamos selecionar três pontos no espaço, e solucionar a equação da reta para esses pontos.

````python
# Quando o valor da equação da reta for igual a zero, esse 
# ponto estara sobre a reta. 

p1 = (2,0.4)
ret1 = a*p1[0] + b*p1[1] + c

plot_line(a,b,c,x)
plt.plot(p1[0],p1[1],color = 'b',marker= 'o')
print('p1 = %.2f'%(ret1))
````

````python
# Quando o valor da equação da reta for menor que zero, esse 
# ponto estara a baixo da reta. 

p2 = (3,0.4)
ret2 = a*p2[0] + b*p2[1] + c

plot_line(a,b,c,x)
plt.plot(p2[0],p2[1],color = 'r',marker= 'o')
print('p2 = %.2f'%(ret2))
````

````python
# Quando o valor da equação da reta for maior que zero, esse 
# ponto estara a baixo da reta. 

p3 = (1,1)
ret3 = a*p3[0] + b*p3[1] + c

plot_line(a,b,c,x)
plt.plot(p3[0],p3[1],color = 'g',marker= 'o')
print('p3 = %.2f'%(ret3))
````

Note que o ponto **p1** está na reta, quanto que o **p2** está a baixo da reta e **p3** esta acima. Ao solucionar a equação da reta para esses três pontos, tivemos respectivamente resultados **nulo (= 0)**, **positivo (> 0)**, **negativo (< 0)**.

Esse comportamente pe consistente para quaisquer pontos na reta ou abaixo dela. Ou seja, se chamarmos a equação da reta *f(x)*, temos as seguintes regras:
+ f(x) = 0 define pontos da reta;
+ f(x) < 0 define pontos abaixo da reta;
+ f(x) > 0 define pontos acima da reta.

## Modelo Linear
Nossa reta pode ser interpretada como um modelo linear, qua nada mais é do que uma função de mapeamento X ➡ Y, que mapeia cada ponto em X para um ponto em Y. Em outras palavras, dados os parâmetros W = {w1,w2} e b de uma reta é possivel mapear uma entrada X = {x1,x2} para a saida f(x;W,b).

Para problemas de classificação os valores de **y** para novas entradas **x** vão definir se **x** é um ponto acima ou abaixo da reta, formando portanto um classificador capaz de separar linearmente problemas com duas classes.

## Regra Geral
Formalizando a nomenclatura de redes neurais, em duas dimensões *(x1,x2)* nosso modelo linear tem dois pesos *(w1,w2)* e um viês **b**, que em inglês são chamados de **w**eight e **b**ias.

Em duas dimensões nosso modelo forma uma reta, como vimos anteriormente. Para um número de dimensões *d > 2*, modelos lineares são chamados de hiperplanos, e são compostos por:

+ um peso *wi* para cada dimensão *xi*.
+ um unico viês *b*.

Exemplo, para 3 dimenções *(x1,x2,x3)* teriamos três pesos *(w1,w2,w3)* e um unico viés b. Sua função de mapeamento seria *y = (w0 x0) + (x1 x1) + (w2 x2) + b*.

## Exercicio Sugerido
**Vamos treinar um classificador linear em duas dimensões usando a tecnologia mais avançada que existe: o seu cérebro!**
A célula a seguir produz uma distribuição aleatória para um problema de classificação com duas classes, usando a função *make_classification()* do sklearn.
Documentação: [make_classification](https://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_classification.html)

````python
from sklearn.datasets import make_classification
import numpy as np

# Fixando a raiz de aleatorização dos dados
np.random.seed(46)

X, Y = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
)
#X: Dimenções dos dados.
#Y: Rotulos.
print('X: ',X.shape)
print('Y: ',Y.shape)
````

````python
# Plotando as distribuições
import matplotlib.pyplot as plt
plt.scatter(X[:,0],X[:,1],marker='o',c=Y, edgecolor='k')
````

````python
# Plotar um ponto especifico
plt.scatter(X[:,0],X[:,1],marker='o',c=Y, edgecolor='k')
p = X[10]
print('Rotulo = ',Y[10])
plt.plot(p[0],p[1], marker='^', markersize=10, color='r')
````

**Seu trabalho** é encontrar os melhores parametros (W, b) de um modelo linear para classificar a distribuição gerada. O ideal é não fazer  por tentativa e erro, mas prestarm atenção no impacto de cada mudança realizada. Note que os pesos W são componentes angulares do seu modelo, enquanto o viés *b* é o componente linear que desloca a rede no espaço.

````python
# Modelo de classificação de separação dos modelos no espaço

def plot_model(w1,w2,b,raiz = 46):



  # Fixando a raiz de aleatorização dos dados
  np.random.seed(46)

  # Gerando a distribuição
  X, Y = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
  )
  plt.scatter(X[:,0],X[:,1],marker='o',c=Y, edgecolor='k')

  #Definir os limites de cada eixo
  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()

  #Gerando o modelo
  x = np.linspace(-2, 4, 50)
  y = (-w1*x -b)/w2

  plt.axvline(0, -1,1,color='k', linewidth=1)
  plt.axhline(0, -2,4,color='k', linewidth=1)
  plt.plot(x,y)
  plt.grid(True)

  #Delimitando
  plt.xlim(xmin,xmax)
  plt.ylim(ymin, ymax)


w1 = 2
w2 = 0.75
b = -0.5
plot_model(w1,w2,b)
````

````python
# Criando a função plotar ponto
def plot_ponto(w1,w2,b,ponto,cor):


  retorno = w1*ponto[0]+w2*ponto[1]+b
  print('Retorno: ',retorno)
  plt.plot(ponto[0],ponto[1], marker='^', markersize=10,color=cor)

w1 = 2
w2 = 0.75
b = -0.5
plot_model(w1,w2,b)
p = (-1,1)
plot_ponto(w1,w2,b,p,'r')
````

Agora podemos utilizar o nosso modelo como classificador linear com simples if para avaliar a solução de equação da reta para novos pontos no espaço.

````python
def classify(w1,w2,b,ponto):
  ret = w1*ponto[0]+w2*ponto[1]+b
  if ret >= 0:
    classe = (1,'y')
  if ret < 0:
    classe = (0,'b')
  return classe

w1 = 2
w2 = 0.75
b = -0.5
p = (2,-1)
c = classify(w1,w2,b,p)
print('Classe: ',c[0])
plot_model(w1,w2,b)
plot_ponto(w1,w2,b,p,c[1])
````

````python
w1 = 2
w2 = 0.75
b = -0.5
p = (-1,1)
c = classify(w1,w2,b,p)
print('Classe: ',c[0])
plot_model(w1,w2,b)
plot_ponto(w1,w2,b,p,c[1])
````

De posse da função classify() podemos agora classificar cada um dos pontos gerados aleatoriamente pelo make_classifcication, calculando assim a acuracia do nosso modelo.

````python
# Valor da acuracia
def acuracia_modelo(w1,w2,b,X,Y):
  cont = 0
  for index in range(len(X)):
    categ,_ = classify(w1,w2,b,X[index])
    if categ == Y[index]:
      cont+=1
  return cont/len(X)


w1 = 2
w2 = 0.75
b = -0.5
print('Acuracia: ',acuracia_modelo(w1,w2,b,X,Y))
````

# Classificação linear com Pythorch
O perceptron é a unidade fundamental de redes neurais, ele recebe *d* conexões de entrada e **produz um único valor de saida**.

Seu processamento é composto por:

+ Função de mapeamento equivalente a regra geral de modelos lineares.

+ Função de ativação *f*.

O trenamento dessa unidade (e de todas as outras em uma rede) busca entrar os parâmetros **W = {w1,w2, ...., wn} e b**. que minimizam o problema em questão.

## Pythorch, criando um modulo nn

* Além da computação tensorial efieicente, o Pythorch também oferece o módelo nn para a construção de redes neurais (**n**eural **n**etwroks).



```
from torch import nn
```

* Atavés desse módelo é possivel instanciar uma ampla diversidade de camadas neurais com chamadas de alto nivel.

* Para instaciar uma camada referente ao perceptron clássico, utilizamos a camada **Linear**. Ela recebe dois parâmetros: 

  * in_features: dimensionalidade da entrada.
  * out_features: dimensionalidade da saida.

* Na representação classicva do perceptron temos uma entrada com 3 dimensões.
  * *X = {x1,x2,x3}*

* E uma unica saída (um numero), produto da transformação linear.
  * *y = WX + b*

* Em termos da camada Linear:
  * in_features: 3
  * out_features: 1


## Aplicação

````python
import torch
from torch import nn
````

````python
# Transformando em parametros numpy
#w1*x1 + w2*x2 + w3*x3 + b = 0
w1, w2, w3 = perceptron.weight.data.numpy()[0]
b = perceptron.bias.data.numpy()
````

É importante não se esquecer que no perceptron esses valores definem um hiperplano no espaço. Como 3 dimensões está no limite da nossa capacidade de visualização, vamos aproveitar esse momento para visualizar o plano aleatoriamente inicializado pelo torch.

````python
# plotando o modelo
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def plot_3d(perceptron):
  W1, W2, W3 = perceptron.weight.data.numpy()[0]
  b = perceptron.bias.data.numpy()

  #Limitando espaço
  X1 = np.linspace(-1,1,10)
  X2 = np.linspace(-1,1,10)

  X1,X2 = np.meshgrid(X1,X2)
  # Obtendo os valores de X3, ou seja, quando X1 for 
  #igual a n1 e X2 for igual a n2 quando valera X3.
  X3 = (b - W1*X1 - W2*X2)/W3

  fig = plt.figure()
  ax = fig.gca(projection='3d')
  #Definir rotação
  ax.view_init(azim=180)

  ax.plot_surface(X1, X2, X3, cmap='plasma')

plot_3d(perceptron)

````

# Funções de Ativação *f*

<Duvida>

De forma abstrada, um neurônio interpreta as suas entradas e libera uma ativação com determinada força.

Esse mesmo neuronio provavelmente fica inativo qunado você bate o dedão em uma quina.

O papel da função de ativação é definir se o neurônio vai ativar e qual a sua função de ativação.
<Duvida>

Quando fizemos o nosso primeiro modelo linear (a reta), a função classify() solucionava a equação da reta um dado ponto e tomava uma decisão.
  
  * if ret >= 0: return 1# Categoria 1
  * else: return 0       # Categoria 0

Essa decisão é uma **função limear**. Uma transformação não lenear que limita a saída do perceptron a 0 ou 1.

  * 0: inativo
  * 1: ativo

Para que o neurônio possa ativar com diferentes "forças", existem funções como a **sigmóide**. Ela também define um intervalo [0,1], mas pode assumir qualquer valor dentro desse intervalo.

Funções de ativação populares:

  * Limiar (pouco utilizada)
  * Sigmoid (Varia entre 0 e 1)
  * ReLU (antes do valor limite o valor sera zero)
  * Tanh (Varia entre 1 e -1)

Graças a elas (entre outras coisas), redes neurais são capazes de solucionar problemas tão complexos com geração de imagens artificiais ou modelos de linguagem.

O módulo nn, que utiliza para criar uma camada linear, tambem comtém as principais funções de ativação. Então vamos primeiro importar o torch e o módulo nn.

````python
import torch
from torch import nn
````

````python
# Modelo de classificação de separação dos modelos no espaço

def plot_model(w1,w2,b,raiz = 46):

  # Fixando a raiz de aleatorização dos dados
  np.random.seed(46)

  # Gerando a distribuição
  X, Y = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
  )
  plt.scatter(X[:,0],X[:,1],marker='o',c=Y, edgecolor='k')

  #Definir os limites de cada eixo
  xmin, xmax = plt.gca().get_xlim()
  ymin, ymax = plt.gca().get_ylim()

  #Gerando o modelo
  x = np.linspace(-2, 4, 50)
  y = (-w1*x -b)/w2

  plt.axvline(0, -1,1,color='k', linewidth=1)
  plt.axhline(0, -2,4,color='k', linewidth=1)
  plt.plot(x,y)
  plt.grid(True)

  #Delimitando
  plt.xlim(xmin,xmax)
  plt.ylim(ymin, ymax)


w1 = 2
w2 = 0.75
b = -0.5
plot_model(w1,w2,b)
````

Agora vamos iunstanciar uma camada Linear análoga à nossa reta, ou seja:
  * in_features = 2 (entrada)
  * outr_features = 1 (saida)

Vamos instanciar tamb´´em uma ativalçai segmoide para compor o nosso modelo de classificação. Documentação: [Ainda não tem link](https://).

Em seguida vamos substituir os pesos do perceptron com os parámetros da nossa reta. Para isso criamos um objetivo Parameter, o tipo de dados Documentação: [Ainda não tem link](https://).

````python
perceptron = nn.Linear(2, 1)
activation = nn.Sigmoid()

print('weight:', perceptron.weight.data)
print('bias:',perceptron.bias.data)
````

````python
#Inserir os dados dentro do perceptron
w1 = 2
w2 = 0.75
b = -0.5

perceptron.weight = nn.Parameter(torch.Tensor([[w1, w2]]))
perceptron.bias   = nn.Parameter(torch.Tensor([b]))

print('weight:',perceptron.weight.data)
print('bias:',perceptron.bias.data)
````

Temos portanto:
  * uma distribuição de pontos com duas classes linearmenet separaveis.
  * um modelo linear de classificação binaria, empiricamente definido, que inclui uma função de ativação sigmóide.

Agora vamos ver a diferença entre o resultado apresentado pelo perceptron antes e depois da função de ativação. Para isso, selecionei empiricamente alguns pontos de distrivuição (indices 17,43, e 66) que nos ajudarão a entendfer o impacto da sigmóide.

````python
# Fixando a raiz de aleatorização dos dados
np.random.seed(46)

# Gerando a distribuição
X, Y = make_classification(
    n_features=2, n_redundant=0, n_informative=1, n_clusters_per_class=1
    )
````
